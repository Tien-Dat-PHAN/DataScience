1. Create brownian noise
2. Rank transform
3. Scale to (-1+1)
4. Apply inverse hyperbolic tangent

The goal of this task is to transform a chaotics signal into a Gaussian distribution signal. It's very appicable in reality

```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
```

```python
n = 500
x = np.cumsum(np.random.randn(n))
y = stats.rankdata(x)/(n+1)*2-1
print(np.min(y), np.max(y))
y = np.arctanh(y)   #remarable step
plt.plot(y)
```

```python
fig, ax = plt.subplots(2,2,figsize=(12,8))
ax[0,0].plot(x)
ax[0,0].set_title('Original data')
ax[0,1].plot(y)
ax[0,1].set_title('Transformed data')

ax[1,0].hist(x, bins = 40)
ax[1,0].set_title('Original data')
ax[1,1].hist(y, bins = 40)
ax[1,1].set_title('Transformed data')
```
