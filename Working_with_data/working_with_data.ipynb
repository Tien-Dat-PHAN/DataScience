```python
# the simplest case is when you have one-dimensional data set, which is just a collection of numbers.
# an ovious step is to computr a few summary statistics . you would like to know how many data points you have, the smallest, the largest, the mean, and the standard deviation 
# then, a good next step is to create a histogram, in which you group your data into discrete buckets and count how many points 
```

```python
import random as rd
import math
from collections import Counter
from matplotlib import pyplot as plt
import scipy.stats as spi

def bucketize(point, bucket_size):
    return bucket_size * math.floor(point/bucket_size)

def make_histogram(points, bucket_size):
    '''bucket the points and counts how many in each bucket'''
    return Counter(bucketize(point, bucket_size) for point in points)

def plot_histogram(points, bucket_size, titile = ''):
    histogram= make_histogram(points,bucket_size)
    plt.bar(histogram.keys(), histogram.values(), width = bucket_size)
    plt.title(titile)
    plt.show()

```

```python
# run an example
import numpy as np
rd.seed(0)

# uniform between -100 and 100
uniform = [200* rd.random() - 100 for i in range(10000) ]


# normal distribution mean = 0, deviation = 57

normal = [np.random.normal(0, 57) for i in range(10000)]

spi.ttest_ind(uniform, normal)
```

```python
plot_histogram(uniform, 10, 'Uniform Histogram')
```

```python
plot_histogram(normal, 10, 'Normal Histogram')
```

```python
plt.hist(uniform, 100)
plt.show()
```

```python
plt.hist(normal, 100)
plt.show()
```

```python
# now work with two dimensions
def random_normal():
    return np.random.normal(0, 57) 

xs = [random_normal() for _ in range(10000)]
ys1 = [ x+ random_normal()/2 for x in xs]
ys2 = [ -x + random_normal()/2 for x in xs]


```

```python
plot_histogram(ys1, 10, 'ys1')
```

```python
plot_histogram(ys2, 10, 'ys2')
```

```python
# it can be similar histogram but each has a very differnent joint distribution with xs,

plt.scatter(xs, ys1, marker='.', color = 'black', label = 'ys1')
plt.scatter(xs, ys2, marker='.', color = 'blue', label = 'ys2')
plt.xlabel('xs')
plt.ylabel('ys')
plt.legend()
plt.show()
```

```python
# check p-value from ttest of scipy.stats
from scipy.stats import ttest_ind

t_test, pval = ttest_ind(ys1, ys2, axis = 0)
print(t_test, pval)
```

```python
# A p-value larger than a chosen threshold (eg 5% or 1%) indicates that our observation is not so unlikely to have occurred by chance. Therefore, we do not reject the null hypothese of equal population mean
# if the p-value is smaller than our threshold, then we have evidence against the null hypothese of equal population means

  
```
